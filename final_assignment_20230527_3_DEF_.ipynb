```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
%reload_ext google.colab.data_table

df = pd.read_csv("/content/drive/MyDrive/levelized-cost-of-energy-b.csv", sep=";", skiprows=1) #read in csv file
df
```

```python
df.drop('Code', axis=1, inplace=True)
df
```

```python
df.columns = ['Country','Year','Bioenergy', 'Geothermal', 'Offshore_wind', 'Solar_photovoltaic', 'Concentrated_solar_power', 'Hydropower', 'Onshore_wind']
df
```

```python
df1 = df[(df['Country'] == 'World') & (df['Year'] > 2011)]
df1.set_index('Year', inplace=True)
df1
```

```python
from sklearn.linear_model import LinearRegression
df1
x = df1.index
x1 = df1.index
y1 = df1['Bioenergy']
y1
x2 = df1.index
y2 = df1['Geothermal']
y2
x3 = df1.index
y3 = df1['Offshore_wind']
y3
x4 = df1.index
y4 = df1['Solar_photovoltaic']
y4
x5 = df1.index
y5 = df1['Concentrated_solar_power']
y5
x6 = df1.index
y6 = df1['Hydropower']
y6
x7 = df1.index
y7 = df1['Onshore_wind']
y7

#plt.figure(figsize=(12,6))

m, b = np.polyfit(x1, y1, 1) #obtain m (slope) and b(intercept) of linear regression line
plt.plot(x1, m*x1+b) #add linear regression line to scatterplot 
m, b = np.polyfit(x2, y2, 1) #obtain m (slope) and b(intercept) of linear regression line
plt.plot(x2, m*x2+b) #add linear regression line to scatterplot 
m, b = np.polyfit(x3, y3, 1) #obtain m (slope) and b(intercept) of linear regression line
plt.plot(x3, m*x3+b) #add linear regression line to scatterplot 
m, b = np.polyfit(x4, y4, 1) #obtain m (slope) and b(intercept) of linear regression line
plt.plot(x4, m*x4+b) #add linear regression line to scatterplot 
m, b = np.polyfit(x5, y5, 1) #obtain m (slope) and b(intercept) of linear regression line
plt.plot(x5, m*x5+b) #add linear regression line to scatterplot 
m, b = np.polyfit(x6, y6, 1) #obtain m (slope) and b(intercept) of linear regression line
plt.plot(x6, m*x6+b) #add linear regression line to scatterplot 
m, b = np.polyfit(x7, y7, 1) #obtain m (slope) and b(intercept) of linear regression line
plt.plot(x7, m*x7+b) #add linear regression line to scatterplot 

plt.scatter(x, y1, c ="r", linewidths = 2, alpha=0.5, label = 'Bioenergy')
plt.scatter(x, y2, c ="y", linewidths = 2, alpha=0.5, label = 'Geothermal')
plt.scatter(x, y3, c ="b", linewidths = 2, alpha=0.5, label = 'Offshore wind')
plt.scatter(x, y4, c ="g", linewidths = 2, alpha=0.5, label = 'Solar photovoltaic')
plt.scatter(x, y5, c ="c", linewidths = 2, alpha=0.5, label = 'Concentrated solar power')
plt.scatter(x, y6, c ="m", linewidths = 2, alpha=0.5, label = 'Hydropower')
plt.scatter(x, y7, c ="c", linewidths = 2, alpha=0.5, label = 'Onshore wind')
#plt.scatter(x, y2, c ="k", linewidths = 2,marker ="p",edgecolor ="red",s = 150, alpha=0.5)
plt.axhline(y=0.05, color='k', linestyle='dotted')
plt.title('Energy prices for nonfossil energy\n', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Price per unit of energy generated\n(US dollars per kilowatt-hour)')
plt.grid()
plt.legend(bbox_to_anchor=(1.5, 1), loc='upper right')

plt.show()
```

```python
from scipy.stats import linregress
plt.figure(figsize=(12,6))

y = df1['Bioenergy']
x = df1.index

plt.scatter(x, y,  c ="g", linewidths = 2, alpha=0.5, label = 'Bioenergy levelized cost of energy')

years_extended = np.arange(2021, 2040, 1)
slope, intercept, r_value, p_value, std_err = linregress(x, y)
line = [slope*xi + intercept for xi in years_extended]

plt.plot(years_extended, line, color = 'g', linestyle='dotted', label="Linear regression line prediction", linewidth=1)

m, b = np.polyfit(x, y, 1) #obtain m (slope) and b(intercept) of linear regression line
plt.plot(x, m*x+b, color = 'g', linestyle='solid', label = "Linear regression line") #add linear regression line to scatterplot 

plt.xticks(range(2012, 2041, 2))
plt.gca().set_ylim(bottom=0)
plt.axhline(y=0.05, color='k', linestyle='dashdot', label = 'Fossil fuels - low end of the price range')
plt.title('Energy price for Bioenergy\n', fontsize=20, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Price per unit of energy generated\n(US dollars per kilowatt-hour)')
plt.grid()

plt.legend()
plt.show()
```
